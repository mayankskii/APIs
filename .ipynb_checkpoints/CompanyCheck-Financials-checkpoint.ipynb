```python
def TAE(expr): # Trial and Error Module
    try:
        var = eval(expr)
    except:
        var = None
    return var
```

```python
# Pending Task: Apply Failsafe
def CompanyFinancials(soup):
    if soup!=None:
        try:
            # Non returned variables
            a1 = TAE('''soup.find_all('tr')''')

            # MidSupport Variables
            if len(a1) > 0:
                keys = []
                for i in range(len(a1)):
                    keys.append(a1[i].th.text)
                financials = []
                for i in range(len(a1)):
                    b = a1[i].find_all('td')
                    values = []
                    for j in range(len(b)):
                        values.append(b[j].text)
                    financials.append(values)
            else: return
                
            dict0 = dict(zip(keys, financials))
            
            a2 = TAE('''soup.find('div', class_ ='data-table-container Financial-account-table' )''')
            if len(a2) > 0:
                PNL_timeline = []
                for i in a2.find_all('time'):
                    PNL_timeline.append(i.text)
            else: return
            
            dict1 = {'PNL_timeline': PNL_timeline}
            
            dict1.update(dict0)
            
            # Returned Variables
            dict2 = {}
            for key, value in dict1.items():
                dict2[key] = value[0]

            # EndTransform Variables

        except (IndexError, KeyError): return
    else: return
    return [dict2, dict1]
```
